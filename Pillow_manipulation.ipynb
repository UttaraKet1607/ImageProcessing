```python
image=image.convert("RGB")
```

```python
blurimg=image.filter(PIL.ImageFilter.BLUR)
sharpimg=image.filter(PIL.ImageFilter.SHARPEN)
edgeimg=image.filter(PIL.ImageFilter.FIND_EDGES)
display(blurimg)
display(sharpimg)
display(edgeimg)
```

```python
medimg=image.filter(PIL.ImageFilter.MedianFilter)
maximg=image.filter(PIL.ImageFilter.MaxFilter)
minimg=image.filter(PIL.ImageFilter.MinFilter)
display(medimg)
display(maximg)
display(minimg)
```

```python
from PIL import ImageEnhance #takes value between 0.0 (complete black) to 1.0(normal image)
```

```python
enhancer=ImageEnhance.Brightness(image)
images=[]
for i in range(0,10):
    images.append(enhancer.enhance(i/10))
print(images)
```

```python
first_img=images[0]
sheet=PIL.Image.new(first_img.mode, (first_img.width,10*first_img.height))
iniloc=0
for img in images:
    sheet.paste(img,(0,iniloc))
    iniloc=iniloc+600
sheet=sheet.resize((160,900))
display(sheet)
```

```python
first_img=images[0]
sheet=PIL.Image.new(first_img.mode, (3*first_img.width,3*first_img.height))
x=0
y=0
for img in images[1:]:
    sheet.paste(img,(x,y))
    if x+first_img.width==sheet.width:
        x=0
        y=y+first_img.height
    else:
        x=x+first_img.width
sheet=sheet.resize((600,600))
display(sheet)
```
