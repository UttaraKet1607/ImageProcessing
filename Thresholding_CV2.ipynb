## Thresholding using cv2

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
```

```python
img = cv2.imread('m.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_OTSU)
```

```python
plt.imshow(img)
```

```python
plt.imshow(thresh)
```
